Now let's work with some real sound files! We'll use various routines from Brian McFee's <a href="https://github.com/bmcfee/librosa">librosa</a>, which is based on work done in Dan Ellis's <a href="http://labrosa.ee.columbia.edu">LabROSA</a>. 

Many of these examples are taken from <a href="http://nbviewer.ipython.org/github/bmcfee/librosa/blob/master/examples/LibROSA%20demo.ipynb">Brian's demo notebook</a>.

```python
# basic imports

import os

# this one lets us call external command line programs
from subprocess import Popen

import numpy as np
# we wanna make plots like in Matlab!
# same idea, we can just type "plt" instead of the whole name
import matplotlib.pyplot as plt

import librosa

import random
import math

%matplotlib inline

```

```python
# Load 60 seconds of a wave file
# By default, librosa will resample the signal to 22050Hz.
# You can change this behavior by setting the sr parameter of librosa.load(), or disable resampling entirely by setting sr=None.

#filename = "MajorLazer-Come_On_To_Me_SeanPaul.mp3"
#filename = "let_it_go.wav"
filename = "ByeByeBlackbird.mp3"
#filename = "Roanoke.mp3"

y, sr = librosa.load(filename, duration=60.0, sr=None)

# show the time series

# make a figure with a 1/3 aspect ratio
plt.figure(figsize=plt.figaspect(.3))
plt.title("waveform")
plt.ylabel("amp")
plt.xlabel("seconds")
# plot it, with the x axis as seconds. What's linspace doing here???
plt.plot(np.linspace(0.0, y.size/sr, y.size), y)

```

```python
# zoom in to the first second, plot in yellow ('y')
plt.figure(figsize=plt.figaspect(.3))
plt.title("one second")
plt.ylabel("amp")
plt.xlabel("seconds")
plt.plot(np.linspace(0.0,1.0,sr), y[:sr], 'y')
```

```python
# zoom in to the middle 100 samples
plt.figure(figsize=plt.figaspect(.3))
plt.title("super zoom!")
plt.ylabel("amp")
plt.xlabel("sample number")
# plot with magenta circles and lines
plt.plot(y[len(y)/2 - 50:len(y)/2 + 50], 'mo-')
```

```python
# Let's make and display a mel-scaled power (energy-squared) spectrogram
# We use a small hop length of 64 here so that the frames line up with the beat tracker example below.
S = librosa.feature.melspectrogram(y, sr=sr, n_fft=2048, hop_length=64, n_mels=128)

# Convert to log scale (dB). We'll use the peak power as reference.
log_S = librosa.logamplitude(S, ref_power=S.max())

# Make a new figure
plt.figure(figsize=(12,4))

# Display the spectrogram on a mel scale
# sample rate and hop length parameters are used to render the time axis
librosa.display.specshow(log_S, sr=sr, hop_length=64, x_axis='time', y_axis='mel')

# Put a descriptive title on the plot
plt.title('mel power spectrogram')

# draw a color bar
plt.colorbar(format='%+02.0f dB')

# Make the figure layout compact
plt.tight_layout()
```

A chromagram does pitch detection and then collapses all of the pitches into an octave, giving you "chroma" or pitch class as the output.

```python
# We'll use a longer FFT window here to better resolve low frequencies
C = librosa.feature.chromagram(y=y, sr=sr, n_fft=4096, hop_length=64)

# Make a new figure
plt.figure(figsize=(12,4))

# Display the chromagram: the energy in each chromatic pitch class as a function of time
# To make sure that the colors span the full range of chroma values, set vmin and vmax
librosa.display.specshow(C, sr=sr, hop_length=64, x_axis='time', y_axis='chroma', vmin=0, vmax=1)

plt.title('Chromagram')
plt.colorbar()

plt.tight_layout()
```

Let's see how many of each pitch class we got. We'll use the mean value as a threshold for how much energy has to be in a bin before we call it that pitch class for sure.

The internet claims that the Demi Lovato version of Let It Go is in G. Let's see what we get...

```python
p_classes = ['a', 'a#', 'b', 'c', 'c#', 'd', 'd#', 'e', 'f', 'f#', 'g', 'g#']

p_class_counts = []

# let's find the average chroma value
# then below we'll use that to decide what counts as a "real" pitch of that class
mean = C.mean()
print "mean is:", mean

for p_class_name, values in zip(p_classes, C):
    # short way
    count = (values > mean).sum()
    # long way
    #count = sum(n > mean for n in p) 

    print p_class_name, "\t", count
    p_class_counts.append(count)

'''
here's a long way to find the pitch class with the most entries:

max_count = max(p_class_counts)
max_count_index = p_class_counts.index(max_count)
key = p_classes[max_count_index]

'''

# and here's a more compact way
key = p_classes[p_class_counts.index(max(p_class_counts))]

print "probable key is: ", key

plt.xticks(np.arange(0,12), p_classes)
plt.bar(np.arange(0,12), p_class_counts, align='center')
```

```python
# Now, let's run the beat tracker
tempo, beats = librosa.beat.beat_track(y=y, sr=sr, hop_length=64)

# Let's re-draw the spectrogram, but this time, overlay the detected beats
plt.figure(figsize=(12,4))
librosa.display.specshow(log_S, sr=sr, hop_length=64, x_axis='time', y_axis='mel')

# Let's draw lines with a drop shadow on the beat events
plt.vlines(beats, 0, log_S.shape[0], colors='k', linestyles='-', linewidth=2.5)
plt.vlines(beats, 0, log_S.shape[0], colors='w', linestyles='-', linewidth=1.5)
plt.axis('tight')

plt.tight_layout()
```

```python
print 'Estimated tempo:        %.2f BPM' % tempo

print 'First 5 beat frames:   ', beats[:5]

# Frame numbers are great and all, but when do those beats occur?
print 'First 5 beat times:    ', librosa.frames_to_time(beats[:5], sr=sr, hop_length=64)


# Events mostly occur at the center of the frame due to Hann windowing.
# Including the n_fft option to frames_to_time adds in the proper correction
beat_times = librosa.frames_to_time(beats, sr=sr, hop_length=64, n_fft=2048)
print 'Corrected beat times:  ', beat_times[:5]

# And at what samples numbers?
# these should be integers! There are no fractional sample numbers...
beat_sample_nums = (beat_times * sr).astype(int)


print 'First 5 sample times:  ', beat_sample_nums[:5]
```

```python
# beat times are nice, how about beat lengths?
# numpy can tell you the difference between each item in an array with np.diff()
# the difference between two beat times gives you the length of that beat!

beat_lengths = np.diff(beat_times)


plt.figure(figsize=plt.figaspect(.3))
plt.title("beat durations")
plt.xlabel("beat number")
plt.ylabel("seconds")
plt.plot(beat_lengths)

print "min: ", beat_lengths.min(), " max:", beat_lengths.max(), "avg:", beat_lengths.mean()

plt.figure(figsize=plt.figaspect(.3))
plt.ylabel("amp & duration")
plt.xlabel("time")
plt.plot(beat_times[:-1], beat_lengths, 'm')
# scale y for major lazer...
plt.plot(np.linspace(0, y.size / sr, y.size), y * 0.75)
```

Maybe we wanna chop up our sample into beats for later use. Let's write a bunch of .wav files.

Why don't we also just put all the beats in separate arrays for easy access.

```python

beat_samples = []

for i in range(beat_sample_nums.size):
    # since we're writing a lot of files, let's put them in a directory
    # make sure to create the directory yourself first!
    beat_filename = "./beats/beat_" + str(i) + "_" + filename + ".wav"
    
    # we'll grab samples from the original file
    
    if i < beat_sample_nums.size - 1:
        samples = y[beat_sample_nums[i]:beat_sample_nums[i+1]]
    else:
        # if it's the last beat we'll just go to the end of the song
        samples = y[beat_sample_nums[i]:]
        
    # and write them to a wave file
    librosa.output.write_wav(beat_filename, samples, sr)
    
    beat_samples.append(samples)
    

print "num beats:", len(beat_samples)
```

And let's make a function that will tell us the filename, given a beat number

```python
def get_filename_for_beat(beat_num, original_filename):
    beat_filename = "./beats/beat_" + str(beat_num) + "_" + original_filename + ".wav"
    
    if os.path.isfile(beat_filename):
        return beat_filename
    else:
        print beat_filename, "doesn't exist!"
        return None
```

how about sparse beats in a longer file?


```python
soundfile_duration = 30.0
num_samples = 2000

# make a large array of zeros. sr * 30 = 30 seconds, right?
sparse_song = np.zeros(int(sr * soundfile_duration))

for i in range(num_samples):
    
    # pick a sample
    the_sample = beat_samples[np.random.randint(0, len(beat_samples))]
    # time in master file where we want sample, in seconds
    # don't go all the way to the end, gotta leave room for the sample
    start_time = np.random.random() * (soundfile_duration - 2.0)
    
    # convert to samples
    start_samp = int(sr * start_time)
    end_samp = start_samp + the_sample.size
    
    sparse_song[start_samp:end_samp] += the_sample



librosa.output.write_wav("sparse_song.wav", sparse_song, sr)

```

how about a normal distribution of beats?

```python
soundfile_duration = 30

# make a large array of zeros. sr * 30 = 30 seconds, right?
normal_dist_song = np.zeros(int(sr * soundfile_duration))

#generate values with a normal (bell-shaped) distribution)
start_times = np.random.randn(len(beat_samples))

# shift them to go from zero to 58 seconds
start_times = np.interp(start_times,[start_times.min(), start_times.max()],[0.0, soundfile_duration - 2.0])
# sort the start times so that we play the beats in order
start_times.sort()

for beat, start_time in zip(beat_samples, start_times):

    # convert to samples
    start_samp = int(sr * start_time)
    end_samp = start_samp + beat.size
    
    normal_dist_song[start_samp:end_samp] += beat

librosa.output.write_wav("normal_dist_song.wav", normal_dist_song, sr)
   
```

A little snipped of code to generate a numpy array with a sinewave in it.<br>
http://milkandtang.com/blog/2013/02/16/making-noise-in-python/
    

```python
def sine(frequency, length, rate):
    length = int(length * rate)
    factor = float(frequency) * (math.pi * 2) / rate
    return np.sin(np.arange(length) * factor)
```

Let's make a file with a blip at each beat

```python
blip_sine = sine(1000, .1, sr)

blip_song = np.array([])

for beat in beat_samples:
    
    # pmake a copy
    the_beat = np.array(beat)
    
    the_beat[:blip_sine.size] += blip_sine
    
    blip_song = np.append(blip_song, the_beat)

librosa.output.write_wav("blipped_song.wav", blip_song, sr)
```
